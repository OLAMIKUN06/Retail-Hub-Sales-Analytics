## Exploratory Data Analysis

```python
# Calculate total sales after discount
# Assuming 'Discount' is in decimal (e.g., 0.10 for 10%)
sales['totalSalesAfterDiscount'] = round(sales['UnitPrice'] * sales['Quantity'] * (1 - sales['Discount']), 2)
print(sales['totalSalesAfterDiscount'].tail())

# Sum it up
total_sales = round(sales['totalSalesAfterDiscount'].sum(), 2)
print("\nTotal Sales:", total_sales)
```

```python
top_products = sales.groupby('ProductID')['totalSalesAfterDiscount'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
top_customers = sales.groupby('CustomerID')['totalSalesAfterDiscount'].sum().sort_values(ascending=False).head(10)
print(top_customers)
```

```python
avg_sales_per_customer = sales.groupby('CustomerID')['totalSalesAfterDiscount'].mean()
print(avg_sales_per_customer.head())
```

```python
sales['totalSalesAfterDiscount'].describe()
```

```python
sns.scatterplot(x='Discount', y='totalSalesAfterDiscount', data=sales)

plt.title('Discount vs Sales After Discount')

# Format x-axis to 2 decimal places
plt.gca().xaxis.set_major_formatter(FormatStrFormatter('%.2f'))

plt.show()
```

## Visualizations

```python
sns.histplot(sales['totalSalesAfterDiscount'], bins=20, kde=True, color='salmon')
plt.title('Sales Distribution')
plt.show()
```

```python
category_sales = sales.merge(products, on='ProductID')
category_sales_grouped = category_sales.groupby('CategoryID')['totalSalesAfterDiscount'].sum()
category_sales_grouped.plot(kind='bar', title='Total Sales by Product Category', color='darkred')
plt.ylabel('Total Sales')
plt.show()
```

```python
sales['SalesDate'] = pd.to_datetime(sales['SalesDate'])
sales['Month'] = sales['SalesDate'].dt.to_period('M')
monthly_sales = sales.groupby('Month')['totalSalesAfterDiscount'].sum()
print(monthly_sales)

monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend (After Discount)')
plt.xlabel('Month')
plt.ylabel('Total Sales After Discount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Months in 'anomalies' may indicate special events, issues, or opportunities.

monthly_sales = sales.groupby('Month')['totalSalesAfterDiscount'].sum()
print(monthly_sales)

# ==============================
# Anomaly detection in monthly sales
# ==============================

# Calculate mean and standard deviation of monthly sales
mean = monthly_sales.mean()
std = monthly_sales.std()

print("\nAverage Monthly Sales:", round(mean, 2))
print("\nStandard Deviation of Monthly Sales:", round(std, 2))

# Define anomalies: months where sales are much higher or lower than usual
# Here we use 2 standard deviations from the mean as the threshold
anomalies = monthly_sales[
    (monthly_sales > mean + 2 * std) | (monthly_sales < mean - 2 * std)
]

print("\nAnomalies (months with unusually high or low sales):")
print('\n',anomalies)
```
